```python
Q1. Explain Class and Object with respect to Object-Oriented Programming. Give a suitable example.
Ans:- In object-oriented programming (OOP), classes and objects are fundamental concepts used to structure and
model software systems. They help organize and encapsulate code, making it more modular, reusable, and easier 
to manage.
    1. Class: A class is a blueprint or template for creating objects. It defines the attributes (data members)
and methods (functions) that objects of that class will have. Think of a class as a set of instructions that 
describe the behavior and properties of a certain type of object.
  For example, let's say we want to create a class to model a basic representation of a "Car." Here's what the 
class might look like in Python:
class Car:
def __init__(self, make, model, year):
    self.make = make
    self.model = model
    self.year = year

def start_engine(self):
    print("Engine started!")

de stop_engine(self):
    print("Engine stopped!")

    2. Object: An object is an instance of a class. It's a concrete realization of the class blueprint, with its
own set of values for the attributes and the ability to perform the methods defined in the class. Objects allow 
you to create multiple instances of the same class, each with its own state and behavior.
  Continuing with the "Car" example, you can create multiple car objects based on the Car class:
# Creating two car objects
car1 = Car("Toyota", "Camry", 2022)
car2 = Car("Ford", "Mustang", 2023)

# Accessing attributes and methods of the objects
print(car1.make, car1.model, car1.year)  # Output: Toyota Camry 2022
car1.start_engine()  # Output: Engine started!

print(car2.make, car2.model, car2.year)  # Output: Ford Mustang 2023
car2.stop_engine()  # Output: Engine stopped!
```

```python
Q2. Name the four pillars of OOPs.
Ans:- The four pillars of Object-Oriented Programming (OOP) are:

1.Encapsulation: Encapsulation refers to the bundling of data (attributes) and methods (functions) that operate
on that data into a single unit called a class. This unit hides the internal details of how the data and methods
are implemented, exposing only the necessary interfaces for interaction. Encapsulation helps in achieving data 
hiding, reducing complexity, and promoting information hiding.
2.Abstraction: Abstraction in Python involves creating classes that provide a simplified interface to the user
while hiding the internal implementation details. You can use abstract base classes (ABCs) and interfaces to 
define a common structure that subclasses should adhere to. Python's native ABC module supports abstraction by
allowing you to create abstract methods that must be implemented by concrete subclasses.
3.Inheritance: Inheritance allows a new class (subclass or derived class) to inherit properties and behaviors
(attributes and methods) from an existing class (superclass or base class). This promotes code reuse and the 
creation of hierarchical relationships between classes. Subclasses can extend or override the functionality of
their parent classes, allowing for specialization while maintaining a consistent structure.
4.Polymorphism: Polymorphism enables objects of different classes to be treated as objects of a common superclass
through a unified interface. This allows for more flexible and dynamic programming, as different objects can respond
to the same method call in ways that are appropriate for their specific types. Polymorphism enhances code reusability
and simplifies code maintenance by promoting interchangeable components.
```

```python
Q3. Explain why the __init__() function is used. Give a suitable example.
Ans:- The __init__() function is used in object-oriented programming to initialize the attributes of an object when it is created from a class. It's like setting up the initial state of the object with the necessary data. This method is automatically called when you create a new instance of a class and allows you to provide values for the object's attributes.
Here's a simple example to illustrate why the __init__() function is used:
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

# Creating instances of the Person class
person1 = Person("Alice", 25)
person2 = Person("Bob", 30)

# Accessing attributes of the objects
print(person1.name)  # Output: Alice
print(person2.age)   # Output: 30
```

```python
Q4. Why self is used in OOPs?
Ans:- In Object-Oriented Programming (OOP) using Python, the self keyword is used to refer to the instance of the class within its methods. It's a convention in Python to use self as the first parameter in method definitions. This parameter allows you to access and manipulate the attributes and methods of the object itself.
    Here's why self is used in Python's OOP:
Instance Reference: When you create an instance (object) of a class, the self parameter represents that specific instance. It allows methods to know which object is calling them and operate on the correct instance's data.

Attribute Access: Inside a method, you use self to access the attributes of the instance. This ensures you're working with the attributes of the specific object that's invoking the method.

Method Invocation: When you call a method on an object, you want it to operate on that specific instance's data. The self parameter ensures that the method interacts with the appropriate data.

Instance Specificity: Different instances of the same class can have different attribute values. self helps maintain this distinction, enabling instance-specific behavior.
```

```python
Q5. What is inheritance? Give an example for each type of inheritance.
Ans:- Inheritance is a fundamental concept in object-oriented programming (OOP) that allows a new class (subclass
or derived class) to inherit properties and behaviors (attributes and methods) from an existing class (superclass
or base class). Inheritance promotes code reuse, hierarchical organization, and the creation of more specialized
classes based on existing ones.
    There are several types of inheritance in Python: 
    1. Single Inheritance: Single inheritance involves a subclass inheriting from a single superclass. The subclass
inherits the attributes and methods of the superclass, and it can also add its own attributes and methods.
class Animal:
    def speak(self):
        pass

class Dog(Animal):
    def speak(self):
        return "Woof!"

class Cat(Animal):
    def speak(self):
        return "Meow!"
    2.Multiple Inheritance: Multiple inheritance allows a subclass to inherit from multiple superclasses. This enables
the subclass to combine attributes and methods from different parent classes.
class Flyable:
    def fly(self):
        pass
class Swimmable:
    def swim(self):
        pass
class FlyingFish(Flyable, Swimmable):
    pass
    3.Multilevel Inheritance: In multilevel inheritance, a subclass inherits from another subclass. This forms a chain
of inheritance, where each class adds its own functionality while building on top of the previous one.
class Vehicle:
    def move(self):
        print("Moving...")
class Car(Vehicle):
    def drive(self):
        print("Car is driving...")
class ElectricCar(Car):
    def charge(self):
        print("Electric car is charging...")
    4.Hierarchical Inheritance: Hierarchical inheritance involves multiple subclasses inheriting from the same superclass.
Each subclass can add its own attributes and methods while sharing the common features of the superclass.
class Shape:
    def draw(self):
        pass

class Circle(Shape):
    def draw(self):
        print("Drawing a circle...")

class Square(Shape):
    def draw(self):
        print("Drawing a square...")

class Triangle(Shape):
    def draw(self):
        print("Drawing a triangle...")
```
